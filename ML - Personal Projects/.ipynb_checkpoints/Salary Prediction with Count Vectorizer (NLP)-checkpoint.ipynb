```python
#use these permutations to input into the system to predict under the model
def permutations(job):
    words = job.lower().split()
    new_words = job.lower().split()
    for i in range(len(words)):
        try:
            new_words.append(words[i] + ' ' + words[i+1])
        except:
            return new_words
    return new_words
```

```python
#function works
permutations('Senior Software Engineer')
```

```python
#using this so just by typing inputs regression can be calculated
def predict_salary_inputs(Age, Gender, Education, Job, Experience, Country, Race, Senior):
    inputs = np.zeros(len(X.columns))
    inputs[0] = Age
    inputs[1] = Experience
    inputs[2] = Senior
    inputs[3] = Education
    if Gender == 'Male':
        inputs[list(X.columns).index('IsMale')] = True
    
    inputs[list(X.columns).index(Race)]= True
    inputs[list(X.columns).index(Country)] = True
    
    job_words = permutations(Job)
    for word in job_words:
        if word in X.columns:
            #print(word, list(X.columns).index(word))
            inputs[list(X.columns).index(word)]= True
    return inputs

def predict_salary(Age, Gender, Education, Job, Experience, Country, Race, Senior):
    return reg.predict([predict_salary_inputs(Age, Gender, Education, Job, Experience, Country, Race, Senior)])[0]
```
